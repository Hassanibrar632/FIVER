```python
#     Parameter

steps = 1
train_size = 0.7
```

```python
import pandas as pd
import numpy as np
from keras.layers import Dense, BatchNormalization, Dropout, LSTM, SimpleRNN, Flatten, CuDNNLSTM
from keras.models import Sequential
import keras
import tensorflow as tf
import matplotlib.pyplot as plt
import random as rand
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("all_data.xlsx", names=[i for i in range(1,21)])
df
```

```python
library = pd.read_excel("Library.xlsx", names=["Values"])
library
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
lb = LabelEncoder()
lb.fit(library["Values"])
library["Values"] = lb.transform(library["Values"])
for i in df.columns:
  df[i] = lb.transform(df[i])
df
```

```python
model = Sequential()
model.add(Dense(2048, activation="relu", input_shape=(1220,)))
model.add(Dropout(0.2))
model.add(Dense(1080, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(1024, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(512, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(256, activation="relu"))
model.add(Dense(20, activation="relu"))

opt = tf.keras.optimizers.Adam(learning_rate=0.0001, decay=0.00001)

model.compile(loss="MSE", optimizer="adam")
model.build()
model.summary()
```

```python
result = {}
temp = model
bestmin = float("inf")
best_prediction = []
for steps in range(1, 10):
  
  model = Sequential()
  model.add(Dense(2048, activation="relu", input_shape=(1220,)))
  model.add(Dropout(0.2))
  model.add(Dense(1080, activation="relu"))
  model.add(Dropout(0.2))
  model.add(Dense(1024, activation="relu"))
  model.add(Dropout(0.2))
  model.add(Dense(512, activation="relu"))
  model.add(Dropout(0.2))
  model.add(Dense(256, activation="relu"))
  model.add(Dense(20, activation="relu"))

  opt = tf.keras.optimizers.Adam(learning_rate=0.0001, decay=0.00001)

  model.compile(loss="MSE", optimizer="adam")
  model.build()
  model.summary()
  
  X = np.array(df)
  
  train = []
  test = []
  te = []

  for i in range(steps,9146):
    test.append(X[i, :])
    train.append(X[i-steps:i, :])
  test = np.array(test)
  train = np.array(train)
  
  for i in train:
    temp = np.zeros(1220)
    for j in i:
     for k in j:
        temp[k] += 1
    te.append(temp)
  te = np.array(te)

  train=te

  x_train, x_test, y_train, y_test = train_test_split(train, test, train_size=train_size)

  history = model.fit(x_train, y_train, epochs=100, validation_split=0.1)

  plt.plot(history.history["loss"], color="blue")
  plt.plot(history.history["val_loss"], color="green")
  plt.legend(["Training LOSS", "Validating LOSS"])
  plt.show()

  predict = model.predict(x_test)
  loss = model.evaluate(x_test, y_test)
  result[steps] = loss
  if loss < bestmin:
    bestmin, best_prediction = loss, predict

  plt.plot([max(i) for i in y_test[0:100]], color='red')
  plt.plot([max(i) for i in predict[0:100]], color='blue')
  plt.title("RESULT: MAX")
  plt.legend(["ACTUAL LIMIT", "PREDICTED LIMIT"])
  plt.show()

  plt.plot([min(i) for i in y_test[0:100]], color='green')
  plt.plot([min(i) for i in predict[0:100]], color='orange')
  plt.title("RESULT: MIN")
  plt.legend(["ACTUAL LIMIT", "PREDICTED LIMIT"])
  plt.show()

  plt.plot([np.mean(i) for i in y_test[0:100]], color='red')
  plt.plot([np.mean(i) for i in predict[0:100]], color='blue')
  plt.title("RESULT: VARIENCE")
  plt.legend(["ACTUAL VARIENCE", "PREDICTED VARIENCE"])
  plt.show()
```
