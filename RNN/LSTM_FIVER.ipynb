```python
#     Parameter

steps = 2
```

```python
import pandas as pd
import numpy as np
from keras.layers import Dense, BatchNormalization, Dropout, LSTM, SimpleRNN, Flatten
from keras.models import Sequential
import keras
import tensorflow
import matplotlib.pyplot as plt
import random as rand
```

```python
df = pd.read_excel("all_data.xlsx", names=[i for i in range(1,21)])
df
```

```python
library = pd.read_excel("Library.xlsx", names=["Values"])
library
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
lb = LabelEncoder()
lb.fit(library["Values"])
library["Values"] = lb.transform(library["Values"])
for i in df.columns:
  df[i] = lb.transform(df[i])
df
```

```python
X = np.array(df)
test = []
for i in range(steps,9146):
  test.append(X[i, :])
test = np.array(test)
test.shape
```

```python
re = []
for i in test:
  temp = np.zeros(1220)
  for j in i:
    temp[j] = 1
  re.append(temp)
re = np.array(re)
re.shape
```

```python
sc = MinMaxScaler(feature_range=(0,1))
df = sc.fit_transform(df)
X = np.array(df)
train = []
for i in range(steps,9146):
  train.append(X[i-steps:i, :])
train = np.array(train)
```

```python
model = Sequential()

model.add(LSTM(units=100, return_sequences=True, input_shape=(train.shape[1],20)))
model.add(LSTM(units=75, return_sequences=True))
model.add(LSTM(units=50, return_sequences=True))
model.add(Flatten())
model.add(Dense(1024, activation="relu"))
model.add(Dense(512, activation="relu"))
model.add(Dense(1220, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.build()
model.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, re, train_size=0.8)
x_train.shape, y_train.shape
```

```python
history = model.fit(x_train, y_train, epochs=100, validation_split=0.1, batch_size=32)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["accuracy"])
plt.title("Step = 5")
plt.ylim(0, 1)
plt.show()
```

```python
predict = model.predict(x_test)
predict.shape
```

```python
model.evaluate(x_test, y_test)
```

```python
plt.plot([max(np.argpartition(i, -20)[-20:]) for i in y_test[0:100]], color='red', label="sum of Real values")
plt.plot([max(np.argpartition(i, -20)[-20:]) for i in predict[0:100]], color='blue', label="sum of predicted values")
plt.title("Model Result")
plt.show()
```
